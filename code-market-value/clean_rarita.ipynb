## Data Preparation

```python
# Processing to all dataframes

for df in all_df:
    df['Player'] = df['Player'].apply(lambda x: fn.clean_names(x), 0)
    fn.pos_to_onehot(df)

salary_2020['Player Name'] = salary_2020['Player Name'].apply(lambda x: fn.clean_names(x), 0)
salary_2021['Player Name'] = salary_2021['Player Name'].apply(lambda x: fn.clean_names(x), 0)
```

```python
# Player age data query

print((l_shooting['Age'][l_shooting.Year == 2020] + l_shooting['Born'][l_shooting.Year == 2020]).describe())
print((l_defense['Age'][l_defense.Year == 2020] + l_defense['Born'][l_defense.Year == 2020]).describe())
print((l_goalkeeping['Age'][l_goalkeeping.Year == 2020] + l_goalkeeping['Born'][l_goalkeeping.Year == 2020]).describe())
print((l_passing['Age'][l_passing.Year == 2020] + l_passing['Born'][l_passing.Year == 2020]).describe())

```

```python
# Compare players between years

for df in all_df:
    df_2020 = df['Player'][df.Year == 2020]
    df_2021 = df['Player'][df.Year == 2021]

    leaving = df_2020[~df_2020.isin(df_2021)]
    entering = df_2021[~df_2021.isin(df_2020)]

    print(f'{df=}'.split('=')[0])
    print(f"{len(df_2020)} total players in 2020, {len(df_2021)} total players in 2021")
    print(f"{len(leaving)} players left in 2020, {len(entering)} players joined in 2021.")
```
